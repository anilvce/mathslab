# Matplotlib
matplotlib has an extensive codebase that can be daunting to many new users. However, most of matplotlib can be understood with a fairly simple conceptual framework and knowledge of a few important points.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y, marker='x')
ax1.set_title('SinX')
ax2.plot(x, np.cos(x), marker='o')
ax2.set_title('CosX')
plt.show()
```

```python
%matplotlib inline
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits import mplot3d

%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

def plot_3DSurface(x, y, fx):
    plt.rcParams["figure.figsize"] = [7.00, 3.50]
    plt.rcParams["figure.autolayout"] = True
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    #ax.plot_surface(x, y, fx)
    ax.plot_surface(x, y, fx, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z');
    plt.show()

def plot_3DContour(x, y, fx):
    plt.rcParams["figure.figsize"] = [7.00, 3.50]
    plt.rcParams["figure.autolayout"] = True
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.contour3D(x, y, fx, 50, cmap='binary')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z');
    plt.show()

def f_sinsqrt(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

def f_plane(x, y):
    return 0.12 * x + 0.01 * y + 1.09

x = np.linspace(-10, 10, 10)
y = np.linspace(-10, 10, 10)
x, y = np.meshgrid(x, y)
z = f_plane(x, y)
plot_3DSurface(x, y, z)
z = f_sinsqrt(x, y)
plot_3DSurface(x, y, z)
plot_3DContour(x, y, z)
```

```python
def f_curvsurface(x, y):
    return (3 * x + 23 * y + 11)**2

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
z = f_curvsurface(x, y)
plot_3DSurface(x, y, z)
plot_3DContour(x, y, z)
```

```python
#plot e^x;  x:[0,10]
%matplotlib notebook
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.exp(x)
plt.plot(x, y, label='e^x')
plt.title('e^x')
plt.show()
```

```python
#Plot x**2 * y**2 * z**2 =1
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
z =  np.sqrt(1/(x**2)*(y**2))
plot_3DSurface(x, y, z)
```

```python
#plot sinx * cosy
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
z =  np.sin(x) * np.cos(y)
plot_3DSurface(x, y, z)
plot_3DContour(x, y, z)
```

```python
#plot x^2 * y^3 ; x:[-1,1]; y:[0,3]
x = np.linspace(-1, 1, 100)
y = np.linspace(0, 3, 100)
x, y = np.meshgrid(x, y)
z =  (x**2) * (y**3)
plot_3DSurface(x, y, z)
plot_3DContour(x, y, z)
```

```python
#plot Arcsin(xy)
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
x, y = np.meshgrid(x, y)
z =  np.arcsin(x*y)
plot_3DSurface(x, y, z)
plot_3DContour(x, y, z)
```
